```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Archivo.xlsx', sheet_name=[0,1,2,3,4,5])
```

### Analisis de minas de gr√°fito y la luz que produce

```python
df2B5 = df[0]
dfB5 = df[1]
dfHB5 = df[2]
df2H5 = df[3]
df2B7 = df[4]
dfHB7 = df[5]

df2B5.head(5)    # Mina 2B 0.5
```

```python
#@title Flujo luminoso en funci√≥n del tiempo

N2B5 = len(df2B5)
NB5 = len(dfB5)
NHB5 = len(dfHB5)
N2H5 = len(df2H5)
N2B7 = len(df2B7)
NHB7 = len(dfHB7)

t2B5 = np.arange(0, N2B5*0.5, 0.5)
tB5 = np.arange(0, NB5*0.5, 0.5)
tHB5 = np.arange(0, NHB5*0.5, 0.5)
t2H5 = np.arange(0, N2H5*0.5, 0.5)
t2B7 = np.arange(0, N2B7*0.5, 0.5)
tHB7 = np.arange(0, NHB7*0.5, 0.5)

fig = plt.figure(figsize=(10, 8))

#plt.title(f'Flujo luminoso en funci√≥n del tiempo aplicando una corriente inicial de $4.95 \ [A]$ \n', fontsize=15)

plt.plot(t2B5, df2B5['F'], '>-', label='2B 0.5', markersize=4)
plt.plot(tB5,  dfB5 ['F'], '*-', label='B 0.5',  markersize=4)
plt.plot(tHB5, dfHB5['F'], 'o-', label='HB 0.5', markersize=4)
plt.plot(t2H5, df2H5['F'], '+-', label='2H 0.5', markersize=4)
plt.plot(t2B7, df2B7['F'], '--', label='2B 0.7', markersize=4)
plt.plot(tHB7, dfHB7['F'], '--', label='HB 0.7', markersize=4)

plt.xlabel(f'Tiempo $t \ [s]$', fontsize=12)
plt.ylabel(f'Flujo luminoso $F(t) \ [lm]$', fontsize=12)
plt.title("Flujo luminoso")
plt.legend()
plt.grid()
plt.show()
```

### Eficacia luminosa:

La eficacia luminosa es una medida de qu√© tan bien una fuente de luz produce luz visible.

```python
#@title Eficacia luminosa en funci√≥n del tiempo

El2B5 = df2B5['F']/df2B5['P']
ElB5 = dfB5['F']/dfB5['P']
ElHB5 = dfHB5['F']/dfHB5['P']
El2H5 = df2H5['F']/df2H5['P']
El2B7 = df2B7['F']/df2B7['P']
ElHB7 = dfHB7['F']/dfHB7['P']

fig4 = plt.figure(figsize=(8,6))

#plt.title(f'Eficacia luminosa en funci√≥n del tiempo: $\eta(t) = F/P$ \n', fontsize=17)

plt.plot(t2B5, El2B5, '>-', label='2B 0.5', markersize=4)
plt.plot(tB5,  ElB5 , '*-', label='B 0.5',  markersize=4)
plt.plot(tHB5, ElHB5, 'o-', label='HB 0.5', markersize=4)
plt.plot(t2H5, El2H5, '+-', label='2H 0.5', markersize=4)
plt.plot(t2B7, El2B7, '--', label='2B 0.7', markersize=4)
plt.plot(tHB7, ElHB7, '--', label='HB 0.7', markersize=4)

plt.xlabel(f'Tiempo $t \ [s]$', fontsize=12)
plt.ylabel(f'Eficacia luminosa $\eta(t) \ [lm/W]$', fontsize=12)
plt.title("Eficacia luminosa")
plt.legend()
plt.grid()
plt.show()
```

```python
#@title Energ√≠a disipada en funci√≥n del tiempo

E2B5 = df2B5['P']*t2B5
EB5  =  dfB5['P']*tB5 
EHB5 = dfHB5['P']*tHB5
E2H5 = df2H5['P']*t2H5
E2B7 = df2B7['P']*t2B7
EHB7 = dfHB7['P']*tHB7

fig5 = plt.figure(figsize=(8,6))

#plt.title(f'Energ√≠a disipada en funci√≥n del tiempo: $\epsilon(t) = P*t$ \n', fontsize=17)

plt.plot(t2B5, E2B5, '>-', label='2B 0.5', markersize=4)
plt.plot(tB5,  EB5 , '*-', label='B 0.5',  markersize=4)
plt.plot(tHB5, EHB5, 'o-', label='HB 0.5', markersize=4)
plt.plot(t2H5, E2H5, '+-', label='2H 0.5', markersize=4)
plt.plot(t2B7, E2B7, '--', label='2B 0.7', markersize=4)
plt.plot(tHB7, EHB7, '--', label='HB 0.7', markersize=4)

plt.xlabel(f'Tiempo $t \ [s]$', fontsize=12)
plt.ylabel(f'Energ√≠a disipada $\epsilon(t) \ [W*t]$', fontsize=12)
plt.title("Energ√≠a disipada")
plt.legend()
plt.grid()
```
